```polyglot-notebook
using System;

public class Plant
{
    public string Name;     
    public string Type;     
    public float Height;

    public Plant(string name, string type, float height)
    {
        Name = name;
        Type = type;
        Height = height;
    }

    public virtual void Blossom()
    {
        Console.WriteLine($"Растение {Name} цветет");
    }

    public void Big(float big)
    {
        Height += big;
        Console.WriteLine($"Растение {Name} выросло на {big} см. Текущая высота: {Height} см");
    }
}

public class Flower : Plant
{
    public string Color;         
    public int Count; 

     public Flower(string name, float height, string color, int count) 
        : base(name, "Цветок", height) 
         {
        Color = color;
        Count = count;
         }

     public override void Blossom()
     {
        Console.WriteLine($"Цветок {Name} распускается и радует глаз! Его цвет: {Color}, Количество лепестков: {Count}");
     }
     public void Wither()
    {
        Console.WriteLine($"Цветок {Name} начинает увядать.");
    }
}

public class Tree : Plant
{
    public int Age;            
    public string LeavesType;

    public Tree(string name, float height, int age, string leavesType) 
        : base(name, "Дерево", height)
    {
        Age = age;
        LeavesType = leavesType;
    }

     public override void Blossom()
    {
        Console.WriteLine($"Дерево {Name} покрывается новыми листьями типа: '{LeavesType}'!");
    }

    public void Old(int years)
    {
        Age += years;
        Console.WriteLine($"Дерево {Name} стало старше на {years} лет. Текущий возраст: {Age} лет");
    }
}

Console.WriteLine("РАСТЕНИЯ В САДУ\n");

Flower rose = new Flower("Роза", 32.3f, "Белый", 8);
Tree bzv = new Tree("Береза", 390.0f, 13, "Широколиственный");

Console.WriteLine($"Создан цветок: {rose.Name}, высота: {rose.Height} см");
Console.WriteLine($"Создано дерево: {bzv.Name}, высота: {bzv.Height} см");
Console.WriteLine();

Console.WriteLine("2.Рост растений:");
rose.Big(5.2f);
bzv.Big(30.5f);
Console.WriteLine();

Console.WriteLine("Уникальность цветения:");

Plant[] plants = { rose, bzv };

foreach (var plant in plants)
{
    Console.Write($"{plant.Type} {plant.Name}: ");
    plant.Blossom(); 
}
Console.WriteLine();

rose.Wither();
bzv.Old(6);
Console.WriteLine();


```
